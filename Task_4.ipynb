```python
import markovify

data = """
Artificial Intelligence is changing the world rapidly.
AI is used in healthcare to detect diseases early.
Machine learning helps systems learn from data automatically.
Deep learning is used in image recognition and speech processing.
Technology is improving day by day with AI advancements.
AI is also used in finance, automation, and robotics.
Many companies are investing in artificial intelligence.
AI can solve complex real world problems efficiently.
"""

model = markovify.Text(data)

topic = input("Enter any topic: ")

print("\nGenerated Text:\n")

count = 0
attempt = 0

while count < 3 and attempt < 10:
    sentence = model.make_sentence()
    
    if sentence:
        print("- " + topic + " " + sentence)
        count += 1
    
    attempt += 1

if count == 0:
    print("- " + topic + " AI is growing rapidly in today's world.")
    print("- " + topic + " Machine learning is very useful in real life.")
```
